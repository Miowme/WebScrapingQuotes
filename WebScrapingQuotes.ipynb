# Using BS4 for HTML Parsing

```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "http://www.values.com/inspirational-quotes"
r = requests.get(URL)
print(r.content)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
quotes_list=[]  # a list to store quotes 
table = soup.find('div', attrs = {'id':'all_quotes'}) 

for i in table.children:
    new_data = {}
    # print(i.findNext('img')['alt'].split('#')[0].strip())
    try:
        new_data['quotes'] = i.find_next('img')['alt'].split('#')[0].strip()
        new_data['topic'] = i.find_next('h5').text
    except Exception as e:
        continue
    quotes_list.append(new_data)
```

```python
df = pd.DataFrame(quotes_list)
df.head()
```

```python
wc = WordCloud(width=800, height=400,background_color="white",max_words=1000).generate(' '.join(df.loc[df['topic']=='LEARNING','quotes']))
plt.imshow(wc)
```
